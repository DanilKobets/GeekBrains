````python
import pandas as pd
import seaborn as sns
df = pd.read_csv('sample_data/california_housing_test.csv')
````

````python
sns.scatterplot(data=df, x="households", y="population")
````

````python
sns.relplot(x="longitude", y="median_house_value", kind="line", data=df, aspect=4)
````

````python
sns.relplot(x="bill_length_mm", y="bill_depth_mm", data=penguins, style='species')
````

````python
sns.relplot(x="bill_length_mm", y="body_mass_g", data=penguins, hue='island')
````

````python
sns.relplot(x="bill_length_mm", y="flipper_length_mm", data=penguins, size='sex')
````

````python
penguins.columns
````

````python
cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
g = sns.PairGrid(penguins[cols])
g.map(sns.scatterplot)
````

````python
a = penguins.columns
g = sns.PairGrid(penguins[a])
g.map(sns.scatterplot)
g.map_diag(sns.histplot)
````

````python
penguins.loc[(penguins['bill_length_mm'] > 42), 'new'] = "high"
penguins.loc[((penguins['bill_length_mm'] >= 35) & (penguins['bill_length_mm'] <= 42)), 'new'] = "middle"
penguins.loc[(penguins['bill_length_mm'] < 35), 'new'] = "low"
````
